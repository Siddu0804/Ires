```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
iris = pd.read_csv('IRIS.csv')

# Filter numeric columns
numeric_iris = iris.select_dtypes(include=[float, int])

# Compute correlation matrix
corr_matrix = numeric_iris.corr()

# Plot heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, cmap='cubehelix', linewidths=1, linecolor='k', square=True, mask=False, vmin=-1, vmax=1, cbar_kws={"orientation": "vertical"}, cbar=True)
plt.show()
```
